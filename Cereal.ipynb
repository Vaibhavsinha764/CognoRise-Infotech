```python
df1 = df[['name','sugars','rating','vitamin']]
df1.sort_values(by='sugars',ascending=False, inplace=True)
```

```python
df1.head(10)
```

```python
df1.head(10).plot(
    kind='bar',
    x='name',
    y='sugars',
    title='Top 10 Cereals by Sugar Content',
    legend=False,
    color='orange'
)
plt.ylabel('Grams of Sugar')
plt.show()
```
